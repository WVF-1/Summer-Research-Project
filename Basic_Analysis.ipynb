```python
# Necessary Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
Land_Data = pd.read_csv("C:/Users/rfull/OneDrive/Documents/Summer Research Project/Landscaping_Data.csv")
```

```python
# Remove the columns of data which are non numeric
Land_Data = Land_Data.drop(Land_Data.columns[[20]], axis = 1)
Land_Data.head()
```

```python
# Exploritory Histogram
sns.histplot(data = Land_Data['Temp_Diff'], bins = 5, kde = True, color = 'blue') #Temp_Diff
plt.show()
```

```python
# Exploritory Histogram
sns.histplot(data = Land_Data['Humidity'], bins = 5, kde = True, color = 'blue') #Humidity
plt.show()
```

```python
# Exploritory Histogram
sns.histplot(data = Land_Data['Cloud_Cover'], bins = 5, kde = True, color = 'blue') #Cloud_Cover
plt.show()
```

```python
# Exploritory Histogram
sns.histplot(data = Land_Data['Shade'], bins = 3, kde = True, color = 'blue') #Shade
plt.show()
```

```python
# Overlayed Histogram
sns.histplot(data = Land_Data['Shade'], bins = 3, kde = True, color = 'yellow', label = 'Shade') #Shade
sns.histplot(data = Land_Data['Rain'], bins = 5, kde = True, color = 'blue', label = 'Rain') #Rain
plt.title('Rain Over Shade')
plt.legend()
plt.show()
```

```python
# Plot of average growth per relative size of plant
sns.barplot(data = Land_Data, x = 'Plant_Size', y = 'Growth', ci = "sd", color = 'limegreen')
plt.title('Average Growth per Relative Size')
plt.show()
```

```python
# Chart of average growth, relative to the size of the plants

# Create the aggregated averages
average_growth = Land_Data.groupby('Plant_Size')['Growth'].mean().reset_index()

# Create a scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(average_growth['Plant_Size'], average_growth['Growth'], 
            s=1000,  # Size of the circles
            c=average_growth['Growth'],  # Color based on growth
            cmap='viridis',  # Colormap for different colors
            alpha=0.6,  # Transparency
            edgecolors='w')  # White edge color for circles
plt.title('Average Growth per Relative Size')
plt.xlabel('Plant Size')
plt.ylabel('Average Growth')
plt.colorbar(label='Growth')  # Add a color bar to indicate growth levels
plt.show()
```
